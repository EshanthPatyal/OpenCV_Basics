```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
#imports
# scale for image replication
# Loading image
img = cv2.imread('lena_color.tif')

scale = [0.5,3]

      #dimensions of original image
height, width, channels = (img.shape)

      #dimensions of new image
new_height, new_width = height * scale[1], width * scale[1]
new_img = np.zeros((new_height, new_width, channels), dtype=np.uint8)

      #replication
for i in range(height):
          for j in range(width):
              for k in range(channels):
                  new_img[i*scale[1]:(i+1)*scale[1], j*scale[1]:(j+1)*scale[1], k] = img[i, j, k]

      # Display the original and enlarged images
cv2_imshow(img)

```

```python
#zoomed image
cv2_imshow(new_img)
```

```python
#doing same for scaling down
new_height, new_width = int(height * scale[0]), int(width * scale[0])
new_img = np.zeros((new_height, new_width, channels), dtype=np.uint8)
for i in range(new_height):
        for j in range(new_width):
            for k in range(channels):
                new_img[i, j, k] = img[int(i/scale[0]), int(j/scale[0]), k]
cv2_imshow(img)
```

```python
#shrunk image
cv2_imshow(new_img)
```

```python
#new dimensions
new_height, new_width = height * 2, width *2


#using interpolation methods from opencv
image_zoomed = cv2.resize(img, dsize=(new_height, new_width), 
      interpolation=cv2.INTER_LINEAR)
cv2_imshow(image_zoomed)
```

```python
new_height, new_width = int(height * 0.5), int(width *0.5)


image_shrunk = cv2.resize(img, dsize=(new_height, new_width), 
      interpolation=cv2.INTER_AREA)
cv2_imshow(image_shrunk)
```
